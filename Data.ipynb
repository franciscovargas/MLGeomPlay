## Data Points as vectors

It is quite common that we extract many measurments from a particular experiment. Say for example we want to describe fruit so we obtain diameter, color, texture and maybe roundness. These measurments (features) can be interpreted as dimensions and packed into a data structure which we call a vector (exactly equivalent to $(x,y,z)$ for location in a 3d space)

\begin{equation}
(diameter, color, roundness, texture) \in \mathbb{R}^{4}
\end{equation}

This notation means that our feture vector is an array with 4 elements who themselves are real numbers.

Lets have a go at looking at some vectors !

## Higher dimensions

When looking at vector spaces with dimensionality higher than 3 it is very dificult to visualize what is going on. We can only percieve 3 dimensions thus it is very difficult to visually analyse more. However there is a great data structure for this called the scatterplot matrix:



```python
iris.keys()
sns.pairplot(iris, hue=" class")
```
